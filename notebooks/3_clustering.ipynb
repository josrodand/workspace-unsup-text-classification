```python
import numpy as np 
import pandas as pd 

import nltk
from nltk.corpus import stopwords
import re
from collections import Counter

from sklearn.decomposition import PCA

from tqdm import tqdm
import pickle

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
def load_vectorizer(name):

    with open(f'../data/sample/vectorizers/{vectorizer_name}', 'rb') as f:
        X_vect = pickle.load(f)
    
    return X_vect


def dimension_reduction(embedding):

    pca = PCA(n_components=2, random_state=42)
    pca_vecs = pca.fit_transform(embedding)
    
    return pca_vecs


def plot_pca(pca_vecs, method):
    plt.figure(figsize=(12, 7))

    plt.title(f"PCA plot from {method}")
    plt.xlabel("X0")
    plt.ylabel("X1")

    sns.scatterplot(x=pca_vecs[:,0], y=pca_vecs[:,1], palette="viridis")
    plt.show()
```

```python
vectorizer_list = [
    'X_tfidf.pkl',
    'X_cls_bert.pkl',
    'X_transformers.pkl',
    'X_d2v.pkl'
]

for vectorizer_name in vectorizer_list:
    print("loading ", vectorizer_name)
    X = load_vectorizer(vectorizer_name)
    X_pca = dimension_reduction(X)
    plot_pca(X_pca, vectorizer_name)



```
